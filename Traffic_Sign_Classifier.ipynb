### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.

#Import OpenCV to do some image pre-processing
import cv2

#The first step is to homogenize lighting by using equalization
# Three methods of equalization are compared

def equalization(image):
    "This function aims to equlize an RGB image"
    #As the image is in RGB, it's a good option to convert the image to HLS
    #Then, equalize the L component and get back to RGB
    #REmember that opencv read the image as BGR
    img_hls = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)
    # equalize the histogram of the Y channel
    img_hls[:,:,1] = cv2.equalizeHist(img_hls[:,:,1])
    # convert the hls image back to RGB format
    return cv2.cvtColor(img_hls, cv2.COLOR_HLS2BGR)

def equalization_gray_scale(image):
    '''This function recives an RGB image as input.
    Returns an equilized gray scale image.
    '''
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    #Divide by 255 in order to normalize [0-1]
    return cv2.equalizeHist(img_gray)/255

def equalization_clahe(image):
    '''Equalization using CLAHE'''
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(4,4))
    return clahe.apply(img_gray)/255

def preprocess(X_data):
    '''Equalize every image on the data set'''
    images = []
    for image in X_data:
        images.append(equalization_clahe(image))
    return np.array(images)


def plot_gallery_eq(images,titles, n_row=3, n_col=6):
    """Helper function to plot a gallery of the images after equalization"""
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=0, right=1.1, top=1)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(equalization_clahe(images[i]), cmap='gray')
        plt.title(titles[i], size=10)
        plt.xticks(())
        plt.yticks(())
        
plot_gallery_eq(images, titles)
plt.show()
````

# Data augmentation

````python
#The classes with less than the average will be augmentated

from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util

def random_rotation(image_array):
    # pick a random degree of rotation between 25% on the left and 25% on the right
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)

def random_noise(image_array):
    # add random noise to the image
    return sk.util.random_noise(image_array)

mean_pic = int(np.mean(distribution)) #around 800 per class

#create dictionary with the class(key) and the desired images(value)
missing_pictures = {}

for i in range(len(distribution)):
    if distribution[i] < mean_pic:
        missing_pictures[i] = mean_pic - distribution[i]

X_to_add = []
Y_to_add = []
for class_index in missing_pictures.keys():
    #Index of the pictures with i class
    pictures_index = np.where(y_train == class_index)
    for i in range(missing_pictures[class_index]):
        random_pic = random.choice(pictures_index[0])
        image = X_train[random_pic].squeeze()
        random_filter = random.randint(0,1)
        if random_filter == 0:
            X_to_add.append(random_rotation(image))
        else:
            X_to_add.append(random_noise(image))
        Y_to_add.append(class_index)

X_train = np.append(X_train, np.array(X_to_add)[..., newaxis], axis=0)
y_train = np.append(y_train, np.array(Y_to_add), axis=0)

print('New traning set shape: X ',X_train.shape)

````

### Model Architecture

````python
### Define your architecture here.

from tensorflow.contrib.layers import flatten

def LeNet(x):
        
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # TODO: Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    F_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma))
    F_b = tf.Variable(tf.zeros(6))
    strides = [1, 1, 1, 1]
    padding = 'VALID'
    conv1 = tf.nn.conv2d(x, F_W, strides, padding) + F_b

    # TODO: Activation.
    conv1 = tf.nn.relu(conv1)

    # TODO: Pooling. Input = 28x28x6. Output = 14x14x6.
    
    ksize = [1, 2, 2, 1]
    strides = [1, 2, 2, 1]
    padding = 'SAME'
    conv1 = tf.nn.avg_pool(conv1, ksize, strides, padding)

    # TODO: Layer 2: Convolutional. Output = 10x10x16.
    F_W_2 = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    F_b_2 = tf.Variable(tf.zeros(16))
    strides = [1, 1, 1, 1]
    padding = 'VALID'
    conv2 = tf.nn.conv2d(conv1, F_W_2, strides, padding) + F_b_2
    
    # TODO: Activation.
    conv2 = tf.nn.relu(conv2)

    # TODO: Pooling. Input = 10x10x16. Output = 5x5x16.
    ksize = [1, 2, 2, 1]
    strides = [1, 2, 2, 1]
    padding = 'SAME'
    conv2 = tf.nn.avg_pool(conv2, ksize, strides, padding)
    
    # TODO: Flatten. Input = 5x5x16. Output = 400
    fc0   = flatten(conv2)
    
    # TODO: Layer 3: Fully Connected. Input = 400. Output = 200.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 200), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(200))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # TODO: Activation.
    fc1    = tf.nn.relu(fc1)
    
    #Dropout
    fc1    = tf.nn.dropout(fc1, keep_prob)

    # TODO: Layer 4: Fully Connected. Input = 200. Output = 90.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(200, 90), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(90))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # TODO: Activation.
    fc2    = tf.nn.relu(fc2)
    
    #Dropout
    fc2    = tf.nn.dropout(fc2, keep_prob)

    # TODO: Layer 5: Fully Connected. Input = 90. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(90, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b

    
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

# Training

````python
### Train your model here.
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)

one_hot_y = tf.one_hot(y, 43)

rate = 0.004

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.8})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
````
